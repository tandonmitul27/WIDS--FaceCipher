```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Aim is to predict the marks of students of the test data
```

```python
# Use the file named 'training data' to train the model

data = pd.read_excel('Training data.xlsx')
x_train = np.array(data.iloc[:,0:8])
y_train = np.array(data.iloc[:,8]).reshape(-1,1)

# Try plotting y_train with different features
# To get an idea whether to add some features or not
column_names = data.columns
num_features = x_train.shape[1]
for i in range(x_train.shape[1]):
    plt.figure(figsize=(6, 4))
    plt.scatter(x_train[:, i], y_train, alpha=0.69)
    plt.xlabel(f'{column_names[i]}') 
    plt.ylabel('y_train')
    plt.title(f'{column_names[i]} vs y_train')
    plt.show()
    
# Add some features if required in x_train

# Also do label encoding for features not represented in numbers
# refer the link if not know : https://youtu.be/589nCGeWG1w?si=t2Wa7LgbUOO4RooM


def feature_changing(x_train):
        
    x_train[:, 0] = np.where(x_train[:, 0] == 'yes', 0, 1) # yes -> 0 and no -> 1
    x_train[:, 1] = np.where(x_train[:, 1] == 'M', 0, 1) # M -> 0 and F -> 1
        
    return x_train

x_train = feature_changing(x_train)
print(x_train)
```

```python
def z_score(x_train):

    x_mean = np.mean(x_train, axis=0)  
    x_std = np.std(x_train, axis=0)  

    x_train= (x_train - x_mean) / x_std

    return x_train,x_std,x_mean
```

```python
def cost(x_train,y_train,w,b):

    error = np.dot(x_train, w) + b - y_train 
    sq_error = error**2
    loss = np.mean(sq_error)

    return loss
```

```python
def gradient_descent(x_train,y_train,w,b):

    n = len(y_train) # number of training samples
    alpha = 0.01
    
    y_pred = np.dot(x_train, w) + b

    dw = (-2/n)*np.dot(x_train.T, (y_train - y_pred))
    db = (-2/n)*np.sum(y_train - y_pred)  

    w = w - alpha*dw
    b = b - alpha*db
        

    cost = np.mean((y_pred - y_train) ** 2)  # Mean Squared Error cost
    print(f"Cost: {cost}")

    return w,b
```

```python
x_train = x_train.astype(np.float64)
x_train,x_std,x_mean = z_score(x_train)

np.random.seed(396836)
w = np.random.randn(x_train.shape[1],1)
b = np.random.randn(1)

old_cost = 0

while abs(old_cost - cost(x_train,y_train,w,b))>0.000001:
    old_cost = cost(x_train,y_train,w,b)
    w,b = gradient_descent(x_train,y_train,w,b)

x_predict = pd.read_excel('Test data.xlsx').iloc[:,:8].to_numpy()
x_predict = feature_changing(x_predict)

#x_predict = (x_predict - x_mean)/x_std
# changed this line as this seemed to cause problem... Shouldn't we be normalizing the test variable based on their mean and std dev? 
x_predict = x_predict.astype(np.float64)
x_predict,x_predict_std,x_predict_mean = z_score(x_predict)

ans = pd.read_excel('Test data.xlsx').iloc[:,8].to_numpy()

y_predict = np.dot(x_predict,w) + b

accuracy = 0
for dim in range(len(ans)):
    print(f"inside for loop and tolerence = {abs(y_predict[dim]-ans[dim])}")
    if abs(y_predict[dim]-ans[dim])<0.5: # do not change the tolerance as you'll be checked on +- 0.5 error only
        accuracy += 1
        print(f"Accuracy count is {accuracy}")
        
accuracy = round(accuracy*100/len(ans),2)
ok = 'Congratulations' if accuracy>95 else 'Optimization required'
print(f"{ok}, your accuracy is {accuracy}%")
```
